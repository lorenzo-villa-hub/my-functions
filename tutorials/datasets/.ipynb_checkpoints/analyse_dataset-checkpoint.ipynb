# Import a Dataset to analyse it

We want to import a Dataset to analyse the outputs of the previous Si DOS-BS calculations. We could have also analysed it directly in the previous notebook, but this is useful to show how to import an existing Dataset without creating it from scratch.

```python
# import packages
from pynter.data.datasets import Dataset
```

Now we import the dataset of the Si DOS-BS calculation. 

```python
ds = Dataset.from_directory('/nfshome/villa/local-data/tutorials/Si-BS-dataset')
print(ds)
```

```python
# check jobs in dataset adding total energy and charge to displayed properties
ds.jobs_table(properties_to_display=['final_energy'])
```

Now let's plot the bandstructure. We can use the methods in the `vasp.analysis` module. The `plot_dos_bs` method uses the `pymatgen` BS plotters.

```python
from pynter.vasp.analysis import JobAnalysis

# First wer select the BS job
step3 = ds.select_jobs(groups=['3-PBE-BS'])
print(step3)
```

```python
JobAnalysis(step3).plot_dos_bs();
```
