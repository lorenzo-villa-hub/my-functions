# Analyse ternary phase diagram

Analyse `PhaseDiagram` object with `pymatgen` and `pynter`.

```python
from pynter.tools.utils import get_object_from_json
from pymatgen.analysis.phase_diagram import PhaseDiagram
from pynter.phase_diagram.analysis import PDHandler
from pymatgen.core.composition import Composition
```

### Get phase diagram object

```python
pd = get_object_from_json(PhaseDiagram,'../data/PD_Mo-Si-Ti.json')
```

```python
pd
```

### View single entry

```python
e = pd.all_entries[0]
e
```

### Plot

Plots are done wih `pymatgen`. For useful shortcuts use`PDHandler` in `pynter`.

```python
pdh = PDHandler(pd)
```

```python
comp = Composition('Si3Mo5')
ent = pdh.get_entries_from_comp(comp)
pdh.get_plot()
```

### Stability diagram

```python
pdh.get_stability_diagram(['Mo','Ti'])
```
