```python
from pynter.tools.materials_project import MPDatabase
from pynter.data.datasets import Dataset
from pynter.vasp.default_inputs import DefaultInputs
from pynter.vasp.schemes import Schemes
import os.path as op
```

```python
ds = Dataset(path='./files/vacancy-inputs')

bulk_structure = MPDatabase('mp-3136').get_structure(conventional_unit_cell=False)
bulk_structure.make_supercell(2)

charge_states ={'Na':[-1,0,1],
                'Nb':[-5,-4,-3,-2,-1,0],
                 'O':[-2,-1,0,1,2]}
system = 'NN'

for el in bulk_structure.composition.elements:
    structure = bulk_structure.copy()
    for site in bulk_structure.sites:
        if site.specie == el:
            structure.remove_sites([bulk_structure.index(site)])
            break
    di = DefaultInputs(structure)
    vaspinput = di.get_vasp_input(xc='PBE')
    cs = Schemes(f'./files/vacancy-inputs/{el.symbol}-vacancy',vaspinput,
                 name='_'.join([system,f'{el.symbol}-vacancy']))
    charge_jobs = cs.charge_states(charges=charge_states[el.symbol],locpot=True)
    for job in charge_jobs:
        pbe_scheme_jobs = Schemes(job.path,job.inputs,job.job_settings,name=job.name).pbe_rel(scheme_name='')
        ds.add_jobs(pbe_scheme_jobs)
```
