# Create inputs for vacancies calculations with VASP

This is just an example, do not run the notebook!

```python
from pynter.tools.materials_project import MPDatabase
from pynter.data.datasets import Dataset
from pynter.vasp.default_inputs import DefaultInputs
from pynter.vasp.schemes import Schemes
from pymatgen.io.vasp.inputs import Poscar
import os.path as op
```

```python
total_jobs = []

bulk_structure = Poscar.from_file('path/CONTCAR').structure #change to CONTCAR

calc_path = 'path'

charge_states ={'Na':[-1,0,1],
                'Nb':[-5,-4,-3,-2,-1,0],
                 'O':[-2,-1,0,1,2]}
system = 'NN_Pbcm'

job_settings = {'add_automation':'automation_vasp.py --contcar --chgcar --wavecar --error-check'}

for el in bulk_structure.composition.elements:
    structure = bulk_structure.copy()
    for site in bulk_structure.sites:
        if site.specie == el:
            structure.remove_sites([bulk_structure.index(site)])
            break
    di = DefaultInputs(structure)
    vaspinput = di.get_vasp_input(xc='PBE')
    cs = Schemes(path=op.join(calc_path,f'{el.symbol}-vacancy'),vaspinput=vaspinput,
                 job_settings=job_settings,name=system+f'_{el.symbol}-vacancy')
    charge_jobs = cs.charge_states(charges=charge_states[el.symbol],locpot=True)
    for j in charge_jobs:
        pbe_jobs = Schemes(path=j.path,vaspinput=j.inputs,job_settings=j.job_settings,
                           name=j.name).pbe_relaxation(scheme_name='')
        for j in pbe_jobs:
            total_jobs.append(j)
            
ds = Dataset(total_jobs)
ds.jobs_table(properties_to_display=['charge'])
```

```python
step2 = ds.select_jobs(common_node='2-PBE-OPT')
for j in step2:
    j.job_settings['array_size'] = 7
```

```python
ds.write_jobs_input()
```

```python
ds.sync_dataset_to_hpc()
step1 = ds.select_jobs(common_node='1-PBE-SCF')
```

```python
ds.jobs_table(jobs=step1)
```

```python
 for j in step1:
     j.run_job(sync=False,write_input=False)
```
